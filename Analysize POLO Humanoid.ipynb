```python
import csv
import matplotlib.pyplot as plt  
```

```python
timestamps = []
rewards = []


with open('polo_record_multiple_0.tsv') as tsvfile:
  reader = csv.DictReader(tsvfile, dialect='excel-tab')
  for row in reader:
    timestamps.append(float(row['timestamp']))
    rewards.append(float(row['reward']))
```

```python
total_records = 18
```

```python
tsv_files = []

for i in range(total_records):
    tsv_files.append(open('polo_record_multiple_{}.tsv'.format(i)))
```

```python
N_TIMESTAMPS = 3000
```

```python
timestamps = []
rewards = []

for i, f in enumerate(tsv_files):
    reader = csv.DictReader(f, dialect='excel-tab')
    epi_rewards = []
    for row_idx, row in enumerate(reader):
        epi_rewards = [0] * N_TIMESTAMPS
        if i == 0:
            timestamps.append(float(row['timestamp']))
        epi_rewards[row_idx]=float(row['reward'])
    rewards.append(epi_rewards)
```

```python
for f in tsv_files:
    f.close()
```
